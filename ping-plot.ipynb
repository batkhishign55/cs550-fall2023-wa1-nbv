```python
# CS550 Assignment 1
# Team: Batkhishig Dulamsurankhor, Nitin Singh, Vaishnavi Papudesi Babu

import numpy as np
import matplotlib.pyplot as plt

networkLatencyDict = {}

latencyLookup = open('network-test-latency.txt', 'r')

latencyLookupLines = latencyLookup.readlines()

for line in latencyLookupLines:
    dns = line.split(" ")
    networkLatencyDict[dns[0]] = float(dns[1])

fig = plt.figure(figsize = (10, 5))
plt.xticks(rotation='vertical')
plt.bar(list(networkLatencyDict.keys()), list(networkLatencyDict.values()), color ='orange'
)
 
plt.xlabel("DNS")
plt.ylabel("Average RTT")
plt.title("Students enrolled in different courses")
plt.show()
```
